```python
from polygon import RESTClient

client = RESTClient(api_key="zMF_APyBCS2eFxFbptloC7h32V0mW6e9")
```

```python
ticker = "AAPL"

# List Aggregates (Bars)
aggs = []
for a in client.list_aggs(ticker=ticker, multiplier=1, timespan="hour", from_="1999-12-17", to="2023-11-16", limit=50000):
    aggs.append(a)
```

```python
import pandas as pd

# create a dataframe from the aggs list
df = pd.DataFrame(aggs)

# save the dataframe to a CSV file
df.to_csv('AAPL.csv', index=False)
```

```python
df = pd.read_csv('AAPL.csv')

# convert the timestamp column to a datetime
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')

# set the timestamp column as the index
df.set_index('timestamp', inplace=True)

# sort the dataframe by the index
df.sort_index(inplace=True)

df.to_csv('AAPL.csv', index=True)
```
